```python
%load_ext autoreload
%autoreload 2
from bokeh.io import push_notebook,show,output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, OpenURL, TapTool, CustomJS,BoxSelectTool,Rect,Ellipse
import imageio
import numpy as np
from image_vis import *
output_notebook()
```

```python
REDUCE=3
reader = imageio.get_reader('chess.mov')
frame_list=[]
for i,frame in enumerate(reader):
    x = scipy.misc.imresize(frame, (int(frame.shape[0] / REDUCE), int(frame.shape[1] / REDUCE)))
    frame_list.append(x.reshape(1,*x.shape))
    #if i>100:
    #    break
len(frame_list)
video=np.concatenate(frame_list)
del frame_list
```

```python
video.shape
```

```python
X,Y=generate_points_data_from_video(video,num_frames=1000,random_frames=True,im_size=40,num_points=1000,percent_negative_filter=3,dist_max=5)



```

```python
X.shape,Y.shape
```

```python
Y.mean(axis=0)
```

```python
model=build_single_cnn_model(shape=(40,40,3),
                           kernel_size = 10, 
                            pool_size_1 = 2 ,
                             pool_size_2 = 2 ,
                            conv_depth_1 = 8 ,
                            conv_depth_2 = 16 ,
                            drop_prob_1 = 0.5 ,
                            hidden_size = 100 ,fully_connected_padding='valid',
                           fully_connected_kernel=(10,10),
                           final_strides=(1,1),
                           num_classes=2,out_flat=True)
model.summary()
```

```python
80/5
```

```python
model.fit(X, Y,                # Train the model using the training set...
          batch_size=100, epochs=1,
          verbose=1, validation_split=0.1)
```
